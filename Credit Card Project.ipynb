```python
## Question :3.Calculate the city wise spend on each product on yearly basis. Also include a graph representation for the same?
```

```python
# Merge spend data with city info from acquisition
spend_cities = Spend.merge(Customer_acq[['Customer', 'City']], on='Customer', how='left')
```

```python
spend_cities
```

```python
#  Convert Month column to datetime and extract year
spend_cities['Month'] = pd.to_datetime(spend_cities['Month'], format='%d-%b-%y')
spend_cities['Year'] = spend_cities['Month'].dt.year
```

```python
spend_cities['Year']
```

```python
spend_cities['Month']
```

```python
# 4. Group by city, product type, and year and calculate total spend
grouped = spend_cities.groupby(['City', 'Type', 'Year'])['Amount'].sum().reset_index()
```

```python
grouped
```

```python
# 5. Pivot the data for easy plotting: rows = (City, Year), columns = Product Type
pivot = grouped.pivot_table(index=['City', 'Year'], columns='Type', values='Amount', fill_value=0)
```

```python
pivot
```

```python
# 6. Plot results for each city as bars (products as different bars per year)

for city in pivot.index.get_level_values('City').unique():
    city_data = pivot.loc[city]
    city_data.plot(kind='bar', figsize=(10,6))
    plt.title(f'Yearly Spend by Product in {city}')
    plt.xlabel('Year')
    plt.ylabel('Total Spend')
    plt.legend(title='Product Type')
    plt.tight_layout()
    plt.show()
```

```python
pivot.to_csv('city_product_yearly_spend_summary.csv')
```

```python
### Question:  4(a). Create graphs for Monthly comparison of total spends, city wise?
```

```python
# Merge city info
spend_city = Spend.merge(Customer_acq[['Customer', 'City']], on='Customer', how='left')
spend_city['Month'] = pd.to_datetime(spend_city['Month'], format='%d-%b-%y')

# Group by city and month
monthly_city = spend_city.groupby(['City', spend_city['Month'].dt.to_period('M')])['Amount'].sum().unstack(fill_value=0)

# Plot each city's monthly trend
monthly_city.T.plot(figsize=(12, 6))
plt.title('Monthly Comparison of Total Spends, City Wise')
plt.xlabel('Month')
plt.ylabel('Total Spend')
plt.legend(title='City')
plt.tight_layout()
plt.show()
```

```python
### Question:4(b) Comparison of yearly spend on air tickets?
```

```python
# Extract year
spend_city['Year'] = spend_city['Month'].dt.year

# Filter for 'AIR TICKET' transactions
air_ticket = spend_city[spend_city['Type'] == 'AIR TICKET']
yearly_air = air_ticket.groupby('Year')['Amount'].sum()

# Plot
yearly_air.plot(kind='bar', figsize=(8,5))
plt.title('Yearly Spend on Air Tickets')
plt.xlabel('Year')
plt.ylabel('Air Ticket Spend')
plt.tight_layout()
plt.show()
```

```python
### Question:4(c) Comparison of monthly spend for each product (look for any seasonality that exists in terms of spend
```

```python
# Group by product and month
product_monthly = spend_city.groupby(['Type', spend_city['Month'].dt.to_period('M')])['Amount'].sum().unstack(fill_value=0)

# Plot each product's monthly trend
product_monthly.T.plot(figsize=(14, 6))
plt.title('Monthly Spend for Each Product (Seasonality Check)')
plt.xlabel('Month')
plt.ylabel('Total Spend')
plt.legend(title='Product')
plt.tight_layout()
plt.show()
```

```python
### Question 5. Write user defined PYTHON function to perform the following analysis:
##               You need to find top 10 customers for each city in terms of their repayment amount by 
##               different products and by different time periods i.e. year or month. The user should be able 
##               to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the 
##               function should automatically take these inputs while identifying the top 10 customers
```

```python
## Step: 1  Filter Customers by City and Product
selected_city = 'BANGALORE'
selected_product = 'Gold'
acq_filtered = Customer_acq[( Customer_acq['City'] == selected_city) & ( Customer_acq['Product'] == selected_product)]
```

```python
acq_filtered
```

```python
## Step:2 Get Customer List from Filtered Acquisition Data
```

```python
filtered_customers = acq_filtered['Customer'].unique()
```

```python
filtered_customers 
```

```python
## Step:3 Filter Repayment Data for Only Those Customers
```

```python
repay_filtered = Repayment[Repayment['Customer'].isin(filtered_customers)].copy()
```

```python
repay_filtered
```

```python
## Step:4 Add Period (Year or Month) Column
```

```python
repay_filtered['Month'] = pd.to_datetime(repay_filtered['Month'], format='%d-%b-%y')
# You can choose 'year' or 'month' below
period_type = 'year'  # Change to 'month' if you want monthly
if period_type == 'year':
    repay_filtered['Period'] = repay_filtered['Month'].dt.year
else:
    repay_filtered['Period'] = repay_filtered['Month'].dt.to_period('M')
```

```python
period_type
```

```python
## Step 5: Group and Sum Repayments by Customer and Period

summary = (
    repay_filtered
    .groupby(['Customer', 'Period'])['Amount']
    .sum()
    .reset_index()
)
```

```python
summary
```

```python
## Question:6 For Each Period, Get Top 10 Customers?
summary_sorted = summary.sort_values(['Period', 'Amount'], ascending=[True, False])
top_customers = summary_sorted.groupby('Period').head(10)
```
